```python
import csv
```

```python
file = open('guns.csv','r')
csvreader = csv.reader(file)
```

```python
data = list(csvreader)
print(data[:5])
```

```python
headers = data[0]
data = data[1:]
print(data[:5])
```

```python
years = [row[1] for row in data]
```

```python
print(years)
```

```python
years_count = {}
```

```python
for year in years:
    if year in years_count:
        years_count[year] = years_count[year] + 1
    else:
        years_count[year] = 1
print(years_count)
```

```python
import datetime
dates = [datetime.datetime(year = int(row[1]),month = int(row[2]),day =1) for row in data]
print(dates[:5])
```

```python
date_count = {}
for date in dates:
    if date in date_count:
        date_count[date] = date_count[date]+ 1 
    else:
        date_count[date] = 1
print(date_count)
```

```python
sex_counts = {}
race_counts ={}
for row in data:
    if row[5] in sex_counts:
        sex_counts[row[5]] = sex_counts[row[5]] + 1
    else:
        sex_counts[row[5]] = 1
    
    if row[7] in race_counts:
        race_counts[row[7]] = race_counts[row[7]] + 1
    else:
        race_counts[row[7]] = 1
print(sex_counts)
print(race_counts)
```
